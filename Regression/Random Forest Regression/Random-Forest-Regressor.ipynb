# Random Forest Regressor 

It works on ensemble learning.

For the selected value of n_estimator that is the number of trees (for ex N), N number of decision trees are generated for different subsets of the data. For the prediction of value, the predicted value of all the trees are taken and then the average is found that average value is considered to be the predicted value. 

## Importing the libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

## Traning the Random Forest Regressor model on complete dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10,random_state=0)
regressor.fit(X,y)
```

## Visualising the Decision Tree Regression results (higher resolution)

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Predicting a new result

```python
regressor.predict([[6.5]])
```
